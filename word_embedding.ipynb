```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent = ["the glass of milk",
       "the glass of juice",
        "the mug of coffee",
        "I am a developer",
        "I build things",
        "I understand things"]
```

```python
sent
```

```python
vocab_size = 10000
```

```python
one_hot_rep =[one_hot(words,vocab_size)for words in sent]
print(one_hot_rep)
```

```python
### Word Embeddings ###
from tensorflow.keras.layers import Embedding
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
import numpy as np
sent_length = 8
```

```python
embed_layer = pad_sequences(one_hot_rep, padding = 'pre', maxlen = sent_length)
print(embed_layer)
```

```python
dims = 15
```

```python
model = Sequential()
model.add(Embedding(vocab_size,15,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
print(model.predict(embed_layer))
```
